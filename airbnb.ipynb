```python
# now replacing the null values from neighbourhood_group acc to the latitude and longitude
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
import time


# Initialize geolocator
geolocator = Nominatim(user_agent="geoapi")

# Function to get neighbourhood group (e.g., borough like Manhattan, Brooklyn, etc.)
def get_neighbourhood_group(lat, lon):
    try:
        location = geolocator.reverse((lat, lon), timeout=10)
        address = location.raw.get('address', {})
        # Common field that holds the borough name
        return address.get('borough') or address.get('suburb') or address.get('city_district') or 'Not Found'
    except GeocoderTimedOut:
        time.sleep(1)
        return get_neighbourhood_group(lat, lon)
    except Exception as e:
        return 'Error'

# Fill missing neighbourhood_group
df['neighbourhood_group'] = df.apply(
    lambda row: get_neighbourhood_group(row['lat'], row['long']) if pd.isna(row['neighbourhood_group']) else row['neighbourhood_group'],
    axis=1
)
```

```python
# now checking for null values in neighbourhood_group
df['neighbourhood_group'].isnull().sum()
```

```python
#now removing the null values from neighbourhood column,latidude,logitude
df.dropna(subset =['neighbourhood','lat','long'],inplace=True)
```

```python
df.isnull().sum()
```

```python
# deleting country,country_code column as ths data is of new york only so we don't reqired country column for analysis
df.drop(['country','country_code'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
# Replacing values in instant_bookable column with unconfirmed
df['instant_bookable'] = df['instant_bookable'].fillna("Unconfirmed")
```

```python
# replacing the true with yes and false with no in instant_bookable column
df['instant_bookable'] = df['instant_bookable'].astype(str).replace({"True": "Yes", "False": "No"})
```

```python
df.isnull().sum()
```

```python
# Replacing nan values in cancellation_policy column
df['cancellation_policy'] = df['cancellation_policy'].fillna("unknown")
```

```python
# dropping nan values in construction year
df.dropna(subset=['construction_year'],inplace=True)
```

```python
# changing type of construction_year column
df['construction_year'] = df['construction_year'].astype(np.int16)
```

```python
df.isnull().sum()
```

```python
# removing the dollar symbol from price column
df['price'] = df['price'].str.lstrip('$')
```

```python
# changing the dtype of price column
df['price'] = pd.to_numeric(df['price'], errors='coerce').astype('float32')
```

```python
# filling the null values with mean of price column and again changing the dtype to int16
df['price'] = df['price'].fillna(df['price'].mean()).astype('int16')
```

```python
df['price'].isnull().sum()
```

```python
# removing the dollar symbol from service_fee column
df['service_fee'] = df['service_fee'].str.lstrip('$')
```

```python
# changing the dtype of service_fee column
df['service_fee'] = pd.to_numeric(df['service_fee'], errors='coerce').astype('float32')
```

```python
# filling the null values with mean of service_fee column and again changing the dtype to int 16
df['service_fee'] = df['service_fee'].fillna(df['service_fee'].mean()).astype('int16')
```

```python
df['service_fee'].isnull().sum()
```

```python
#dropping the rows in which price is less than service fee
df.drop(index=df[df['price']<df['service_fee']].index,inplace=True)
```

```python
# now changing the column names
df.rename(columns = {'price':'price_in_$','service_fee':'service_fee_in_$'},inplace=True)
```

```python
df.isnull().sum()
```

```python
# filling the missing values in minimum_nights with mean of minimum_nights column and changing the dtype of the column
df['minimum_nights'] = df['minimum_nights'].fillna(df['minimum_nights'].mean()).round().astype(np.int8)
```

```python
df.sample(30)
```

```python
 # fillling nan values with 0 in number_of _reviews column
df['number_of_reviews'] = df['number_of_reviews'].fillna(0).astype(np.int16)
```

```python
# changing the datatype of last review column
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
# filling the reviews_per_month column with 0 where number_of_reviews are 0 
df['reviews_per_month'] = df['reviews_per_month'].mask(df['number_of_reviews']==0,0)
```

```python
# changing the datatype of reviews_per_month column from float64 to float 16
df['reviews_per_month'] = df['reviews_per_month'].astype('float16')
```

```python
# removing the null values from reviews_per_month column
df.dropna(subset=['reviews_per_month'],axis=0,inplace=True)
```

```python
df.head()
```

```python
# filling nan values in review_rate_number with mean value and changing the dtype to int8 to reduce memory usage
df['review_rate_number'] = df['review_rate_number'].fillna(df['review_rate_number'].mean().round()).astype('int8')
```

```python
# filling nan values in calculated_host_listings_count with mean value changing the dtype to int8 to reduce memory usage
df['calculated_host_listings_count'] = df['calculated_host_listings_count'].fillna(df['calculated_host_listings_count'].mean().round()).astype('int8')
```

```python
# converting the days into 365 where the number of days in availability_365 is greater than 365 because a year cannot have more than 365 days
df.loc[df['availability_365']>365,'availability_365']=365
```

```python
# replacing the availability_365 column with mean where availability_365 is 0
df['availability_365'] = df['availability_365'].mask(df['availability_365']==0,df['availability_365'].mean().round())
```

```python
# replacing the availability_365 column with mean where availability_365 is nan
df['availability_365'] = df['availability_365'].fillna(df['availability_365'].mean().round()).astype('int16')
```

```python
# dropping house_rules and license column
df.drop(columns=['house_rules','license'],inplace=True)
```
